```python
# 패키지 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta
import warnings
warnings.filterwarnings('ignore')

# 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/Proj_WT/DataSets/airpolution/WholeData.csv', parse_dates = ['측정일자'])
df.head()
```

# 결측치 제거

```python
# 날짜가 없는 데이터들을 만들어 주기 위해서 측정일자를 다시 만들어준다.
a = 1
data_list = []
data_list.append(pd.to_datetime('20100101'))
for i in range(4017):
  # 시작일부터 1일씩 더해주며 series를 만들어 준다.
  start = pd.to_datetime('20100101')
  start = start + timedelta(days = a)
  a += 1
  data_list.append(start)

# merge를 위해서 데이터 프레임으로 만들어 준다.
final_data = pd.DataFrame(data_list, columns = ['측정일자'])

# merge를 측정일자를 기준으로 만들어 준다.
final_data = pd.merge(final_data, df, how = 'left', on = '측정일자')
```

```python
# 결측치를 확인해 준다.
final_data.isnull().sum()
```

```python
# 값이 0인 값들을 NaN 값으로 만들어 준다.
for i in final_data.columns:
  final_data.loc[final_data[i] == 0, i] = np.nan
```

```python
# 결측치 확인하기
final_data.isnull().sum()
```
