```python
# 값을 뽑아내기 위해서 year, month를 만든다.
data['year'] = data['측정일자'].dt.year
data['month'] = data['측정일자'].dt.month
```

```python
month = ['01','02','03','04','05','06','07','08','09','10','11','12']
year = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
Gaussian_dict = {}

# 연별로 for문을 돌린다.
for i in year:
  # 달별로 for문을 돌린다.
  for j in month:
    # a라는 값에 각 year_month를 저장한다.
    a = data[(data['year'] == i) & (data['month'] == int(j))]
    # dict에 담는다.
    Gaussian_dict[f'{i}-{j}'] = a
```

```python
# 정규성 검정하기
from scipy.stats import shapiro

# 구별 이름을 columns에 담는다.
columns = data.columns[1:-2]

df_gaussian = pd.DataFrame([])
# 구별로 for문을 돌린다.
for i in columns:
  # Gaussian_dict에 연별 정보를 for문을 돌린다.
  for idx, value in Gaussian_dict.items():
    # null값을 제외한다.
    if value.isnull().sum().sum() == 0:
      # 정규성검정을 한다.
      test_stat, p_val = shapiro(value[i])
      # p_value가 0.05이상인 값들을 뽑는다.
      if p_val > 0.05:
        print(f'{i} & {idx} & {p_val}')
```
