### 핵심용어(Key words)
- 1. 날씨 (Weather)
- 2. 기후 (Climate)
- 3. 관광기후지수 (Tourism Climate Index : TCI)
- 4. 야외관광활동 (Outdoor tourism activities)

### 지금 TCI의 한계점
1. 열쾌적성은 서양의 기후들을 가지고 만든거라 우리 나라에 적용하기에는 무리가 있다.
2. 기후조건에 대한 선호는 지역의 지리나 문화에 따라 달라질 수 있다.

# KTCI(한국형 관광기후지수)

- 1. 열쾌적성(한낮) (최고기온, 최저상대습도) -> Cd
- 2. 열적쾌적성(일평균) (평균기온, 평균상대습도) -> Ca
- 3. 강수(mm) (총 강수량) -> P
- 4. 풍속(m/s) (평균풍속) -> W
- 5. 일조시간(h) (일조시간) -> S

- 원래 TCI = 2(4Cd + Ca +2P + W +2S)

본 연구는 기후별로 나눠서 KTCI를 구해줬다.
- 연중 KTCI = 2(2.59Cd + 2.09Ca + 3.22P + 1.16W +0.95S)


- 봄철 KTCI = 2(2.58Cd + 1.93Ca + 3.04P + 1.31W + 1.14S)
- 여름철 KTCI  = 2(3.07Cd + 1.90Ca + 3.27P + 0.90W + 0.86S)
- 가을철 KTCI = 2(2.37Cd + 2.18Ca + 3.43P + 1.04W + 0.98S)
- 겨울철 KTCI = 2(2.27Cd + 2.39Ca + 3.07P + 1.46W + 0.81S)

- 관광 => 휴식, 피크닉, 걷기(산책, 관람), 자전거타기, 기타 등으로 범주화를 시킨 이후에 데이터를 분석하였다.

- 한계점 : 우리는 관광지가 야외에 있는것과 실내에 있는것을 구분하고 분석을 진행하지 않았다. 그러나 이 연구에서는 야외활동으로 분석을 진행했다.
- 그래서 관광지와 기후에 관련한 논문이 아니라 야외에 있는 관광지나, 아니면 야외에 있는 여가활동으로 한정해서 차이가 있을 수 있다고 생각한다.
